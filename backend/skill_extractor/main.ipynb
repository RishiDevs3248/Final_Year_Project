```python
import re
import pandas as pd
import numpy as np
import spacy

# Load the CSV file
data = pd.read_csv("UpdatedResumeDataSet.csv")

# Predefined Skills (or load from a dataset)
skill_keywords = ['Agile', 'Analytics', 'Angular', 'AWS', 'Azure', 'Big Data', 'CSS', 'Data Analysis', 'Data Science', 
                  'Django', 'Docker', 'ETL', 'Excel', 'Flask', 'Git', 'HTML', 'Java', 'JavaScript', 'JIRA', 'Keras', 'Linux', 
                  'Machine Learning', 'MATLAB', 'MongoDB', 'Node.js', 'PostgreSQL', 'Power BI', 
                  'Python', 'R', 'React', 'SAS', 'Scrum', 'Spark', 'SQL', 'Tableau','TensorFlow']

# Preprocessing: Clean Text
def preprocess_text(text):
    text = re.sub(r"[^a-zA-Z0-9\s,]", "", str(text))  # Ensure text is a string and remove special characters
    text = text.lower()  # Convert to lowercase
    return text

# Apply preprocessing to the 'Resume' column (adjust column name as needed)
data["Cleaned_Resume"] = data["Resume"].apply(preprocess_text)

# Extract Skills for each row
def extract_skills(resume_text):
    return [skill for skill in skill_keywords if skill.lower() in resume_text]

data["Extracted_Skills"] = data["Cleaned_Resume"].apply(extract_skills)

# Print results
print(data[["Resume", "Extracted_Skills"]])

# Optional: Use NLP Libraries for Advanced Extraction
nlp = spacy.load("en_core_web_sm")

def extract_skills_ner(resume_text):
    doc = nlp(resume_text)
    return [ent.text for ent in doc.ents if ent.label_ in ["SKILL", "TECH"]]

data["NER_Skills"] = data["Cleaned_Resume"].apply(extract_skills_ner)

# Print results with NER
print(data[["Resume", "Extracted_Skills", "NER_Skills"]])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# Convert list of skills into a single string
data['Extracted_Skills_Str'] = data['Extracted_Skills'].apply(lambda x: ' '.join(x))

# Now you can use LabelEncoder on the flattened string if needed
data['Extracted_Skills_Encoded'] = le.fit_transform(data['Extracted_Skills_Str'])

print(data[['Extracted_Skills', 'Extracted_Skills_Str', 'Extracted_Skills_Encoded']])
# data

```

```python
x = data.iloc[:,6].values
y = data.iloc[:,0].values
y = le.fit_transform(y)
```

```python
x=np.array(x).reshape(-1,1)
y=np.array(y).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25 , random_state=0)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=60,random_state=0)
x_test
rf.fit(x_train,y_train)
```

```python
y_rf=rf.predict(x_test)
y_rf
```
